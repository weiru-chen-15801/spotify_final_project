## APCOMP209A Spotify Final Project
Preprocess the 1000 songlists.

```python
import numpy as np
import pandas as pd

#create empty dataframe
endDF = pd.DataFrame()

#number of CSVs we want to go through
numCSV = 1000

#go through all CSV and get the first row and append it to another dataframe
for i in range(0, numCSV):
    song_df = pd.read_csv("../Songs/songs{}.csv".format(i))
    head = song_df.iloc[0]    
    print("working on", i,"th CSV now...")
    endDF = endDF.append(head, ignore_index=True)

#save it as CSV
endDF.to_csv("./result.csv")

#preview what it looks like at the end
endDF
```
