### Making Histograms of Audio Features

```python
%matplotlib inline
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import pandas as pd
import time
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)
pd.set_option('display.notebook_repr_html', True)
import seaborn as sns
sns.set_style('whitegrid')
sns.set_context('poster')
```

```python
# Make list to read in features dataframes into 1 big dataframe form 1st 100 playlists
playlist_dfs = []
for i in range(100):
    pl = 'df_playlist_features_'+str(i)
    playlist_dfs.append(pl)
```

```python
# Make 1 big dataframe of Audio features form 1st 100 playlists
big_df =pd.DataFrame()
for df in playlist_dfs:
    next_df=pd.read_csv('dfs_first_100_playlists_features/'+df, index_col=0)
    big_df= pd.concat([big_df , next_df], axis = 0)
```

```python
fig, ax=plt.subplots(1,2, figsize=(14,4))

ax[0].hist(big_df.acousticness,label = 'Acousticness')
ax[0].set_xlabel('Acousticness', fontsize = 14)
ax[0].set_ylabel('Frequency', fontsize = 14)
ax[0].set_yticks(np.arange(0, 3500, 500))
ax[0].set_xticks(np.arange(0, 1.2, 0.2))
ax[0].tick_params(labelsize = 14)
ax[0].set_ylabel('Frequency', fontsize = 14)
ax[0].legend(loc='upper center', fontsize = 14)
ax[0].grid(True, lw=1.75, ls='--', color = 'lightgrey', alpha=0.75)

ax[1].hist(big_df.danceability,label = 'Danceability')
ax[1].set_xlabel('Danceability', fontsize = 14)
ax[1].set_ylabel('Frequency', fontsize = 14)
ax[1].set_yticks(np.arange(0, 1500, 200))
ax[1].set_xticks(np.arange(0, 1.2, 0.2))
ax[1].tick_params(labelsize = 14)
ax[1].set_ylabel('Frequency', fontsize = 14)
ax[1].legend(loc='upper left', fontsize = 14)
ax[1].grid(True, lw=1.75, ls='--', color = 'lightgrey', alpha=0.75)
```
