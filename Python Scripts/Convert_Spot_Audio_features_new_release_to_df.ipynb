## Converting Spotify Audio Feature JSON data to New Release Dataframes

```python
import json
import pandas as pd
import os
```

```python
# make playlists list (1 to 100) named album_nr_1, named album_nr_1 ... 
album_nr_songs = []
for i in range(1,100):
    pl = 'album_nr_'+str(i)+'.json'
    album_nr_songs.append(pl)
```

```python
# Given json data files as album_nr_1, album_nr_2 ...
# read in file, convert to data frame in list of dfs, dfs_pls_raw
dfs_new_release_raw=[]
for pl in album_nr_songs:
    with open(pl) as json_file:
        data = json.load(json_file)
    dfs_new_release_raw.append(pd.DataFrame(data['audio_features']))
```

```python
c= ['id','acousticness', 'danceability', 'speechiness','instrumentalness','liveness', 'energy','valence',
        'duration_ms','key','mode', 'loudness', 
        'tempo', 'time_signature']
```

```python
# Given raw dfs in list dfs_new_release_raw
# create a new list of dfs with selected columns
dfs_new_release_short = []
for df in dfs_new_release_raw:
    df2 = df.copy()
    df2 = df2[c]
    dfs_new_release_short.append(df2)
```

```python
# Adding column song name:
dfs_plus_trunc = []
for i, pl in enumerate(dfs_new_release_short):
    #a_name = 'artist_'+str(i)+'.txt'
    t_name = 'alb_track_names_'+str(i+1)+'.txt'  
    #da = pd.read_csv(a_name, names=['artist'])
    dn = pd.read_csv(t_name, names=['track_name'])
    dfs_plus_trunc.append(pd.concat([dn, pl], axis = 1))
```

```python
#Writing each df in dfs_plus_trunc to a file
for i, df in enumerate(dfs_plus_trunc):
    name = 'df_new_release_features_'+str(i)
    df.to_csv(name)
```
