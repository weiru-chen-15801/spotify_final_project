## Creating a Dataframe of Unpopular Songs ##

```python
import json
import pandas as pd
import numpy
```

```python
# read in playlist dfs to a list dfs
dfs = []
for i in range(100):
    name = 'df_playlist_features_'+str(i)
    df = pd.read_csv(name, index_col=None)
    df = df.drop('Unnamed: 0', axis=1)
    dfs.append(df)
```

```python
# read in unpopular songs df to dataframe df_up
df_up = pd.read_csv('Unpopular_Songs')
```

```python
# Searches each playlist dataframe and extract those rows that contained songs in the unpopular dataframe
# and creates a new data frame 'df_un_pop_songs' that contains audio features for the unpopular songs
df_un_pop_songs = pd.DataFrame()
for pop, id, playlist in zip(df_up['popularity'], df_up['id'], df_up['album']):
    df = dfs[playlist-1]
    df2 = df[df['id']==id]
    df2['popularity']=pop
    df2['playlist']=playlist
    df_un_pop_songs = pd.concat([df_un_pop_songs, df2], axis=0)
    df_un_pop_songs = df_un_pop_songs.reset_index(drop=True)
```
