```python
import json
import pandas as pd
import glob
```

```python
# I have a bunch of files labeled new_album_1.json, new_album_2.json ...The files
# are the result of the bash script nr.sh run twice with init = 1 and then again with init = 50.
# Each contains the json info from 1 new release album downloaded from spotify.
# This script opens each json file (99 files) and extracts info about each album
# and puts it into a dataframe. The album id is the 3rd column. I will use
# This column to extract each track and then its audio features from spotify

album_ids=[]
album_names=[]
album_uris=[]
album_total_tracks=[]
album_artists=[]
album_release_dates=[]
for i in range(1,100):
    name = 'new_album_'+str(i)+'.json'
    with open(name) as json_file:
        data = json.load(json_file)
        A = data['albums']['items'][0]
        album_ids.append(A['id'])
        album_release_dates.append(A['release_date'])
        album_names.append(A['name'])
        album_uris.append(A['uri'])
        album_total_tracks.append(A['total_tracks'])
        album_artists.append(A['artists'][0]['name'])
        
df_new_release_albums = pd.DataFrame({'Album_Name': album_names,
                                      'Artist': album_artists,
                                      'Album_id':album_ids,
                                     'Album_uri':album_uris,
                                      'Total_Tracks':album_total_tracks,
                                     'Release_Date':album_release_dates})
```

```python
# Makes a text file named "albums.txt" that is a list of album id files to be read by gt.sh,
# the bash script that will get track ids for each track on the listed albums.

with open('albums.txt', mode='wt', encoding='utf-8') as myfile:
        for album in df_new_release_albums['Album_id']:
            myfile.write(album +'\n')
    
```
